```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sn
```

```python
# Import breast cancer dataset
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
data.keys()
```

```python
# Checking output classes
print(data['target_names'])
```

```python
# Checking input attributes
print(data['feature_names'])
```

```python
# Construct a dataframe using pandas
df1=pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
# Scaling data before applying PCA
scaling=StandardScaler()
```

```python
# Use fit and transform method
scaling.fit(df1)
Scaled_data=scaling.transform(df1)
```

```python
# Set n_components=3
principal=PCA(n_components=3)
principal.fit(Scaled_data)
x=principal.transform(Scaled_data)
```

```python
# Check the dimensions of data after PCA
print(x.shape)
```

```python
# Checking the values of eigen values
# Proceed by principal components
principal.components_
```

```python
# Plotting the components
plt.figure(figsize=(5,5))
plt.scatter(x[:,0],x[:,1],c=data['target'],cmap='plasma')
plt.xlabel('PC_1')
plt.ylabel('PC_2')
```

```python
# Importing relevant libraries for 3D graph
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(10,10))
```

```python
# Choose projection 3D for creating 3D graph
axis=fig.add_subplot(111,projection='3d')
```

```python
# 3d plot
fig = plt.figure()
axis = fig.add_subplot(111, projection='3d')
axis.scatter(x[:,0], x[:,1], x[:,2], c=list(data['target']), cmap='plasma')
axis.set_xlabel("PC1", fontsize=10)
axis.set_ylabel("PC2", fontsize=10)
axis.set_zlabel("PC3", fontsize=10)
plt.show()
```

```python
# Check how much variance is explained by each principal components
print(principal.explained_variance_ratio_)
```
