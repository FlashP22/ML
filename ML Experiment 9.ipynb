```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("/content/pollution_india_2010 (1).csv",na_values='Null')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data=data.dropna()
```

```python
data_pol=data.groupby('State',as_index=False)[['NO2','PM10','SO2']].agg(np.sum)
```

```python
data_pol.head()
```

```python
def scale(x):
    return (x-np.mean(x))/np.std(x)
data_num=data_pol.drop("State",axis=1)
data_scaled=data_num.apply(scale,axis=1)
```

```python
data_scaled.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data_scaled=np.array(data_scaled)
```

```python
Z=linkage(data_scaled,method="ward")
```

```python
fig, ax = plt.subplots(figsize=(15, 20))
ax=dendrogram(Z,orientation="right",labels=np.array(data_pol['State']),leaf_rotation=30,leaf_font_size=16)
plt.tight_layout()
plt.show()
```
