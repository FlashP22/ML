```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Salary_dataset.csv")
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df.dropna(subset=['YearsExperience', 'Salary'], inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X = df[['YearsExperience']]
y = df[['Salary']]
```

```python
plt.scatter(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)

yfit = model.predict(X_train)
yfit
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, yfit)
plt.show()
```

```python
yfit = model.predict(X_test)
plt.scatter(X_test, y_test)
plt.plot(X_test, yfit)
plt.show()
```

```python
from sklearn import metrics
print("Mean Absolute Error: ",metrics.mean_absolute_error(y_test,yfit))
print("Mean Squared Error: ",metrics.mean_squared_error(y_test,yfit))
```
